```python
U = np.array([1, 0.25, 0])
V = np.array([0, 1, 0.25])
W = np.array([0.25, 0, 1])
origin = np.zeros(3)

fig = plt.figure()
ax  = fig.add_subplot(111, projection='3d')
ax.quiver(*origin,*U, color='r'); ax.text(*U,'U')
ax.quiver(*origin,*V, color='g'); ax.text(*V,'V')
ax.quiver(*origin,*W, color='b'); ax.text(*W,'W')
ax.set_xlabel('X'); ax.set_ylabel('Y'); ax.set_zlabel('Z')
plt.show()
U = np.array([-0.5, 0.5])
V = np.array([1, 1])
W = U + V

origin = np.array([0,0])
fig, ax = plt.subplots()
ax.quiver(*origin,*U, angles='xy',scale_units='xy',scale=1,color='r'); ax.text(*U,'U')
ax.quiver(*origin,*V, angles='xy',scale_units='xy',scale=1,color='b'); ax.text(*V,'V')
ax.quiver(*origin,*W, angles='xy',scale_units='xy',scale=1,color='g'); ax.text(*W,'W=U+V')
ax.set_xlim([-1,2]); ax.set_ylim([-1,2])
ax.set_xlabel('X'); ax.set_ylabel('Y')
plt.grid(); plt.show()
```

```python
def magnitude(v):
    return np.sqrt(np.sum(v**2))

U = np.array([1.5, 1.5, 2.0])
print("Magnitude =", magnitude(U))

# ── Task 6d
def direction_cosines(v):
    return v / magnitude(v)

O = np.array([1.5, 1.5, 2.0])
dc = direction_cosines(O)
print("Direction cosines =", dc)

# ── Task 6e
def scalar_product(a, b):
    return np.sum(a * b)

a = np.array([1.5,  1.5, 2.0])
b = np.array([-1.5, 0.5, 2.0])
print("Dot product =", scalar_product(a, b))   # 3.5

# ── Task 6f
def cross_product(a, b):
    return np.array([
        a[1]*b[2] - a[2]*b[1],
        a[2]*b[0] - a[0]*b[2],
        a[0]*b[1] - a[1]*b[0]
    ])

print("Cross product =", cross_product(a, b))
```
